Verify Install

```python
import tensorflow as tf
print(tf.__version__)
```

Create tensor

```python
a=tf.constant(10)
print(a)
v=tf.constant([1,2,3,4])
print(v)

m=tf.constant([[1,2],[3,4]])
print(m)

t3=tf.constant([[[1,2],[3,4]],[[5,6],[7,8]]])
print(t3)
```

Scalar Addition,Subration,Multiplicaton,Division,ppower

```python
a=tf.constant(24)
b=tf.constant(26)
print(tf.add(a,b))
print(tf.subtract(a,b))
print(tf.multiply(a,b))
print(tf.divide(a,b))
print(tf.pow(a,b))
```

Tensor performance on 2D

```python
a=tf.constant([10,20,30])
b=tf.constant([34,2,7])
print(tf.add(a,b))
print(tf.subtract(a,b))
print(tf.multiply(a,b))
print(tf.divide(a,b))
```

Tensor performance on 3D

```python
a=tf.constant([[10,20],[76,13]])
b=tf.constant([[34,2],[35,5]])
print(tf.add(a,b))
print(tf.subtract(a,b))
print(tf.multiply(a,b))
print(tf.divide(a,b))
```

datatype and shape of tensor

```python
a= tf.constant([10,20,30])
print(a.dtype,a.shape,tf.rank(a))
```

Numpy array to tensor

```python
import numpy as np
np_array=np.array([[1,2,3],[4,5,6]])
print(np_array)
tensor=tf.convert_to_tensor(np_array)
print(tensor)
```

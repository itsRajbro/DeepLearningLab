```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
list=["Hello","I","am","student","of","CSEAIML","from","KIET","in","Ghazibad"]
list
```

```python
attention=np.array([
    [0.40, 0.20, 0.20, 0.20],
    [0.10, 0.50, 0.20, 0.20],
    [0.15, 0.20, 0.45, 0.20],
    [0.10, 0.20, 0.30, 0.40]
])
```

```python
attention.sum(axis=1)
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(attention, annot=True, cmap='Greens', xticklabels=list[:attention.shape[1]], yticklabels=list[:attention.shape[0]],fmt='.2f')
plt.title('Attention heatmap')
plt.xlabel('Key')
plt.ylabel('Query')
plt.tight_layout()
plt.show()
```

```python
for i in range(attention.shape[0]):
    query_word = list[i]
    max_index = np.argmax(attention[i])
    print(f'Query word "{query_word}" attends most to "{list[max_index]}" with score {attention[i, max_index]:.2f}')
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
corpus = [
    "Artificial intelligence is changing the world in many ways.",
"Machine learning helps computers learn from data.",
"Deep learning is a subset of machine learning.",
"Neural networks are inspired by the human brain.",
"A dataset is used to train a machine learning model.",
"Training data should be clean and well organized.",
"A model learns patterns from examples in the dataset.",
"LSTM stands for Long Short Term Memory.",
"LSTM is a special type of recurrent neural network.",
"Recurrent neural networks are useful for sequence data.",
"Text generation is a common application of LSTM models.",
"Natural language processing deals with human language.",
"Words are converted into numbers before model training.",
"Tokenization is the process of splitting text into words.",
"Each word can be mapped to a unique integer.",
"Sequences are created from the tokenized text.",
"Padding is used to make all sequences the same length.",
"The model predicts the next word in a sequence.",
"Loss decreases as the model learns better patterns.",
"Accuracy may improve over multiple training epochs.",
"An epoch means one full pass through the dataset.",
"Batch size controls how many samples are processed together.",
"The optimizer updates model weights during training.",
"Adam is a popular optimizer in deep learning.",
"Categorical crossentropy is often used for text prediction.",
"Embedding layers help represent words in dense vectors.",
"Dense vectors capture semantic relationships between words.",
"The hidden state stores useful information over time.",
"Memory cells allow LSTM to remember past context.",
"Forget gates decide what information should be removed.",
"Input gates decide what new information should be stored.",
"Output gates decide what information should be passed on.",
"A trained model can generate meaningful text sequences.",
"Prediction quality depends on data quality and model design.",
"More training data often improves the performance of the model.",
"Overfitting happens when the model memorizes the training data.",
"Validation data helps check model performance during training.",
"Testing data is used to evaluate the final model.",
"Python is commonly used for machine learning projects.",
"TensorFlow is a popular deep learning framework.",
"Keras provides a simple interface for building neural networks.",
"The tokenizer builds a vocabulary from the training corpus.",
"Vocabulary size affects the complexity of the model.",
"The sequence length determines how much context is used.",
"Longer sequences may capture more meaning in text.",
"Shorter sequences train faster but may lose context.",
"Preprocessing is an important step in any machine learning pipeline.",
"A good corpus should contain meaningful and varied sentences.",
"Consistent grammar helps the model learn better patterns.",
"Text generation can be creative and educational.",
"LSTM models are useful for learning sequential dependencies."
]
print("Number of sentences:", len(corpus))
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(corpus)

word_index = tokenizer.word_index
total_words = len(word_index) + 1
print("Vocabulary size:", total_words)
print("\nWord index:")
print(word_index)
```

```python
input_sequences = []

for line in corpus:
    token_list = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(token_list)):
        n_gram_seq = token_list[:i+1]
        input_sequences.append(n_gram_seq)

print("Total training sequences:", len(input_sequences))
print("\nSome sequences before padding:")
for seq in input_sequences[:10]:
    print(seq)
```

```python
max_seq_len = max(len(seq) for seq in input_sequences)
input_sequences = pad_sequences(input_sequences, maxlen=max_seq_len, padding='pre')

print("Maximum sequence length:", max_seq_len)
print("\nPadded sequences:")
print(input_sequences[:10])
```

```python
X = input_sequences[:, :-1]
y = input_sequences[:, -1]

y = tf.keras.utils.to_categorical(y, num_classes=total_words)

print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

```python
model = Sequential([
    Embedding(input_dim=total_words, output_dim=10, input_length=max_seq_len - 1),
    LSTM(50),
    Dense(total_words, activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(X, y, epochs=100, verbose=1)
```

```python

def generate_text(seed_text, next_words):
    for _ in range(next_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=max_seq_len - 1, padding='pre')

        predicted_probs = model.predict(token_list, verbose=0)
        predicted_index = np.argmax(predicted_probs, axis=-1)[0]

        output_word = ""
        for word, index in tokenizer.word_index.items():
            if index == predicted_index:
                output_word = word
                break

        seed_text += " " + output_word
    return seed_text
```

```python
print(generate_text("LSTM is ", 5))
print(generate_text("machine learning", 2))
print(generate_text("python is", 2))
```
